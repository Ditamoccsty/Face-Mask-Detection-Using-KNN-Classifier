```python
import cv2
import os
import numpy as np
```

```python
folder_name = input('name your folder : ')
folder = os.path.join('C:/Users/ACER/FINAL/database', folder_name)
os.makedirs(folder)
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
cam = cv2.VideoCapture(0)
cv2.namedWindow("Frame")
img_counter = 0
```

```python
with_mask=[]
while True:
    ret, frame = cam.read()
    if not ret:
        print("failed to grab frame")
        break
    cv2.imshow("Frame", frame)

    faces = face_cascade.detectMultiScale(frame)
    for x,y,w,h in faces:
        face = frame[y:y + h, x:x + w]
        print( face.shape )
        face_resize = cv2.resize( face, (100, 100) )
        print( face_resize.shape )
        #face_shape = face_resize.reshape(1,-1)
        cv2.rectangle( frame, (x, y), (x + w, y + h), (255, 0, 255), thickness=2 )

    cv2.imshow("Face_object",frame)
    k = cv2.waitKey(1)
    if k % 256 == 27:
        # ESC pressed
        print("Escape hit, closing...")
        break
    elif k % 256 == 32:
        # SPACE pressed
        #cv2.imwrite("capture.jpg", face_resize)

        #image = cv2.imread('capture.jpg')
        #img = cv2.imread('capture.jpg', cv2.IMREAD_GRAYSCALE

        img_name = "{}.png".format(img_counter)
        file = os.path.join(folder, img_name)
        cv2.imwrite(file, face_resize)
        print("{} written!".format(img_name))
        img_src = cv2.imread("{}".format(file))
        cv2.imshow("Segmentasi", img_src)
        img_counter += 1

cam.release()
cv2.destroyAllWindows()
```
